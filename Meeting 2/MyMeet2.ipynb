```python
# EDIT THIS FUNCTION
def myexp(x,nterms):
    p = 0
    for k in range(0,nterms):
        p += x**k/np.math.factorial(k)
    return p
```

```python
# EDIT THIS FUNCTION
def myerf(x,nterms):
    p = 0
    for k in range(0,nterms):
        p += (-1)**(k)*(x**(2*k+1))/(np.math.factorial(k)*(2*k+1))
    return p
```

```python
# EDIT THIS FUNCTION
def mycos(x,nterms):
    p = 0
    for k in range(0,nterms):
        # Compute the sum p_new = p_old + next term in series
        p += (-1)**k*x**(2*k)/np.math.factorial(2*k)
    return p
```

```python
# EDIT THIS FUNCTION
def mysin(x,nterms):
    p = 0
    for k in range(0,nterms):
        # Compute the sum p_new = p_old + next term in series
        p += (-1)**k*x**(2*k+1)/np.math.factorial(2*k+1)
    return p
```

```python
# RUN THIS CELL
# Initialize the grid and number of terms
x = np.linspace(-1,1,50)
nterms = 3
```

```python
# RUN THIS CELL
y = myexp(-x**2,nterms)
plt.plot(x,y,'r',x,np.exp(-x**2),'b.')
plt.xlabel('$x$',size=15)
plt.ylabel('$y$',size=15);
```

```python
# RUN THIS CELL
y = mycos(x,nterms)
plt.plot(x,y,'r',x,np.cos(x),'b.')
plt.xlabel('$x$',size=15)
plt.ylabel('$y$',size=15);
```

```python
# RUN THIS CELL
y = mysin(x,nterms)
plt.plot(x,y,'r',x,np.sin(x),'b.')
plt.xlabel('$x$',size=15)
plt.ylabel('$y$',size=15);
```
