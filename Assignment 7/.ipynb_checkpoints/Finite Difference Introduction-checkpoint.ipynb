```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,1)
u = np.sin(np.pi*x)
plt.plot(x,u);
```

```python
n = len(u)
h = (x[-1] - x[0])/(n-1)
sc = np.array([-1/2, 0, 1/2])/h #[u[0], u[1], u[2]] ... [u[n-3], u[n-2], u[n-1]]
sleft = np.array([-3/2, 2, -1/2])/h #[u[0],u[1],u[2]]
sright = np.array([-1/2, 0, -3/2])/h #[u[-3],u[-2],u[-1]]
```

```python
up = []
up.append(np.array([u[0],u[1],u[2]]).dot(sleft))
for k in range(1,n-1):
    up.append(np.array([u[k-1],u[k],u[k+1]]).dot(sc))
up.append(np.array([u[-3],u[-2],u[-1]]).dot(sright))
```

```python
x = np.linspace(0,1)
u = np.sin(np.pi*x)
plt.plot(x,u,'b',x,up,'r',x,np.pi*np.cos(np.pi*x),'k:');
```
