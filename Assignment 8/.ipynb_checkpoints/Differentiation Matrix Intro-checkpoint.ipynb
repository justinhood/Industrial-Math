# Differentiation Matrices and Spatial-Temporal Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import spdiags, lil_matrix
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

### Second-Order Accurate Central Difference Matrix for First Derivative

###### Example (1D Transport Equation)

\begin{align*}
u_t(x,t) + cu_x(x,t) &= 0, \qquad x > -1, t > 0,\\
u(-1,t) = 0,\\
u(x,0) = f(x)
\end{align*}

Solve this PDE using the second-order accurate central-difference differentiation matrix for $\frac{\partial u}{\partial x}$. Use $c=1$ for the wave-speed and $f(x) = \exp(-25(x+0.5)^2)$ for the initial condition.

```python
N = 39
x = np.linspace(-1,1,N)
h = (x[-1] - x[0])/(N-1)
e = np.ones(N)
#Construct stencil weights          # -1  1 diagonal indices
stencil = np.array([-e/2.,e/2.]) # [-1/2 1/2] e.g. u'[1] = (u[2] - u[0])/2h
#Centered difference stencils are located on the off-diagonals
diag_array = np.array([-1,1])
D = spdiags(stencil,diag_array,N,N)
D = lil_matrix(D)
#Differentiaion at the boundary is approximated using one-sided stencil
D[0] = np.eye(1,N) # 1*u[0] = 0
D[-1,-3:] = np.array([1/2.,-2.,3./2]) # u'[-1] = (1/2)u[-3]- 2u[-2]+(3/2)u[-1]
D = D/h
```

```python
def func(u,r):
    D = r[0]
    c = r[1]
    # u_t + cu_x = 0 implies u_t = -c u_x which implies u_t = -c D*u
    return -np.array(c*D@u)
```

###### Implement Euler's Method

```python
# Initialize the simulation parameters
uv, tv = [], []
u = (x<0)*np.exp(-25*(x+0.5)**2) # The (x < 0) zeros out the positive part
# Time parameters
dt = 1e-3    # Step-size
t, tf = 0, 1 # Initial, final time
c = 1        # Wave-speed
r = (D,c)    # tuple of differentiation matrix and parameter

# Append the list of solutions and times
uv.append(np.array(u))
tv.append(np.array(t))
# Euler's Method
while t <= tf:
    u += dt*func(u, r)
    t += dt
    uv.append(np.array(u))
    tv.append(np.array(t))
```

```python
# Display Results in a Space-Time (spatial temporal) plot
xx,tt = np.meshgrid(x,tv)
uv = np.array(uv)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(xx,tt,uv,cmap='jet')
ax.set_xlabel('$x$',size=14)
ax.set_ylabel('$t$',size=14)
ax.set_zlabel('$u$',size=14)
ax.view_init(elev=50,azim=-100)
ax.set_xticks([-1,0,1])
ax.set_zticks([0,1]);
```
