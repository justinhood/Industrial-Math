```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fEuler1(t,x): #I am the DE w/o Harvesting
    return .3*(1-x/5)*x

def euler(x0, t0, tf, dt): #Std Euler representation
    t, x = t0, x0
    points = [[t,x]]
    while t <= tf:
        x += dt*fEuler1(t,x)
        t += dt               
        points.append([t,x])
    return np.array(points)
```

## Here we compute an array of solutions over the slope field

```python
t0, tf, dt = 0, 5, 1e-1 #Time from 0 to 5,  in steps of .1. Decrease dt for accuracy at cost of more computations.
xmin, xmax = -10,10 

p1 = euler(1, t0, tf, dt)
plt.figure(figsize=[10,7])
plt.plot(p1[:,0],p1[:,1],'r-',p1[:,0],p1[:,1],'b.') #Plot with initl condition x_0=1

p2 = euler(7, t0, tf, dt)
plt.plot(p2[:,0],p2[:,1],'r-',p2[:,0],p2[:,1],'b.') #Plot with initl condition x_0=7

p3 = euler(-1, t0, tf, dt)
plt.plot(p3[:,0],p3[:,1],'r-',p3[:,0],p3[:,1],'b.') #Plot with initl condition x_0=-1

# Construct Slope Field
npts = int((tf-t0)/dt+1)     # Number of time steps taken
t = np.linspace(t0,tf,30)     # Time Grid
x = np.linspace(xmin,xmax,10)      # Solution Data Grid
# Meshgrid Creates Coordinate Matrices from Coordinate Vectors
tt,xx = np.meshgrid(t,x)
# Slope-field
u = 1           # 1D so u does not vary
v = fEuler1(t,xx)     #MY EULER SLOPE FIELD
# Normalize the arrow lengths
N = np.sqrt(u**2+v**2)
u, v = u/N, v/N
# Construct the Slope-field (Notice any Equilibrium Solutions)
plt.quiver(tt,xx,u,v)
plt.xlabel('$t$',size=15)
plt.ylabel('$x(t)$',size=15);
plt.title("Eulers Method") 
plt.savefig('2b.png') #Save as a PNG
```

```python
def fEuler2(t,x): #Now add Harvesting
    h=.2311
    return .2311*(1-x/1072764)*x-h*x

def euler2(x0, t0, tf, dt):
    t, x = t0, x0
    points = [[t,x]]
    while t <= tf:
        x += dt*fEuler2(t,x)
        t += dt               
        points.append([t,x])
    return np.array(points)
```

## Here we run computations on varied $h$ to analyze the effects

```python
t0, tf, dt = 0, 100, 1e-1 #Time from 0 to 100,  in steps of .1. Decrease dt for accuracy at cost of more computations.
xmin, xmax = 0,900000  

p1 = euler2(900000, t0, tf, dt)
plt.figure(figsize=[10,7])
plt.plot(p1[:,0],p1[:,1],'r-',p1[:,0],p1[:,1],'b.') #Plot with initl condition x_0=900000

# Construct Slope Field
npts = int((tf-t0)/dt+1)     # Number of time steps taken
t = np.linspace(t0,tf,60)     # Time Grid
x = np.linspace(xmin,xmax,10)      # Solution Data Grid
# Meshgrid Creates Coordinate Matrices from Coordinate Vectors
tt,xx = np.meshgrid(t,x)
# Slope-field
u = 1           # 1D so u does not vary
v = fEuler2(t,xx)     #MY EULER SLOPE FIELD
# Normalize the arrow lengths
N = np.sqrt(u**2+v**2)
u, v = u/N, v/N
# Construct the Slope-field (Notice any Equilibrium Solutions)
plt.quiver(tt,xx,u,v)
plt.xlabel('$t$',size=15)
plt.ylabel('$x$',size=15);
plt.title("Eulers Method") 
plt.savefig('4b3.png') #Save as a PNG
```
