### Problem 5 Data

###### Below is a set of nodes, $x$, and the values of two unknown functions, $f$ and $g$, at these corresponding grid points. Find an estimate for the point of intersection of the graphs of these two functions. Note that you need to the find BOTH the $x$ and $y$ values for the point!

```python
x = np.array([0.15839511,0.2247862,0.57336078,0.90902066,1.60394509,1.80681731,2.06397836,2.43692578])
f = np.array([0.9752231,0.95072651,0.71982829,0.43765753,0.07633377,0.03821265,0.01412221,0.0026357 ])
g = np.array([1.08193102,1.06379173,0.8770369,0.59224085,-0.01563539,-0.13283609,-0.21963379,-0.22527554])
h = g-f
print(h)
```

```python
plt.plot(x,f,'r',x,g,'b', x,h , 'g',x,x*0,'k') # Note the itnersection point should be around 1.4
```

```python
#Using h=f-g Inverse interpolate h.
H = np.c_[h].dot(np.ones((1,len(h))))
k = np.arange(len(h))
phi = H**k
ch=np.linalg.solve(phi,x)
plt.plot(h,np.dot(phi,ch),'ro',h,x,'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Collocation');
#Show me the coefficients.
print(ch)
```

Note that the constant coefficient is $9.63776687$, So our polynomial interpolant of $h$ at $0$ will be $9.63776687$. This does not align with what we would predict the output to be, as our polynomial is very oscillatory. We further examine this behavior below, by interpolating across the range of $h$ and seeing the oscillations between nodes of the true data. 

```python
# Construct interpolation grid
xi = np.linspace(-.25,.15)
Xi = np.c_[xi]
phi = Xi**k

yi = np.dot(phi,ch) #<-- Notice recycled coefficients c_k's
plt.plot(h,x,'bo',xi,yi,'r.')
plt.xlabel('h')
plt.ylabel('x')
plt.title('Interpolation');
```

```python
# Construct interpolation grid
xi = 0
Xi = np.c_[xi]
phi = Xi**k

yi = np.dot(phi,ch) #<-- Notice recycled coefficients c_k's
plt.plot(h,x,'bo',xi,yi,'r.')
plt.xlabel('h')
plt.ylabel('x')
plt.title('Interpolation');
print(yi)
```

We see that the supposed $x^*$ value that solves this interpolation is approximately 9.6377, which is well outside the domain of the data. This is clearly not the solution that we want. Next, we will brute force a solution with Lagrange interpolation.

```python
#Lagrange attempt
def p(x, xdat, ydat):
    s=0
    n=1
    d=1
    for i in range(0,len(ydat)):
        for j in range(0,len(ydat)):
            if(i!=j):
                d *= xdat[i]-xdat[j]
                n *= x-xdat[j]
        s+=ydat[i]*n/d
        d=1
        n=1
    return s
            
```

```python
#We know the value is approximately 1.4, so starting at 1.39, we iterate
#till we find a value that is zero for both f and g interpolation
err=1
tol=10e-8
xstar=1.39
while(np.fabs(err)>tol):
    err=p(xstar,x,f)-p(xstar,x,g)
    xstar+=tol
xstar-=tol
print("Critical $x$ value")
print(xstar)
print("Lagrange output for $x^*$")
print(p(xstar,x,f))
```

```python
def grad_descent_regression(x,y,eta, tol, itmax):
    iterror = tol+1
    i=1
    errvec = []
    W = np.random.rand(2)
    X = np.c_[np.ones(len(x)),x]
    
    while (iterror > tol) and (i<itmax):
        W0 = W.copy()
        pred = X.dot(W)
        error = pred-y
        grad = X.T.dot(error)
        W-=eta*grad
        
        iterror = np.sqrt((W-W0).dot(W-W0))
        errvec.append(iterror)
        i += 1
    return W, i-1, errvec
```

```python
# Set the simulation parameters
eta = 3e-2
itmax = 5000
tol = 1e-4
# Run GD
w, iters, errvec = grad_descent_regression(x,y,eta,tol,itmax)
# ******************************************************************
# Display the results
# ******************************************************************
print("Step-size: {:}, tolerance: {:}".format(eta,tol))
print("After {:d} iterations the min is {:}".format(len(errvec),w))
plt.plot(np.arange(len(errvec))+1,errvec)
plt.xlabel('iterations',size=15)
plt.ylabel('error',size=15)
plt.axis([1,len(errvec),0,max(errvec)])
plt.grid(True);
plt.figure()
plt.plot(x,y,'bo',x,w[0] + w[1]*x,'r')
plt.xlabel('$x$',size=25)
plt.ylabel('$y$',size=25)
plt.grid(True);
```
