```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.sparse import spdiags, lil_matrix
%matplotlib inline
```

```python
mu=4.375359e-8
L=20
dP=.0275
a=.3
```

```python
v = lambda r: (np.pi*dP)*r**4/(8*mu*L)
```

```python
r=np.linspace(0,a,100)
vr=v(r)
r=r/a
#r=np.flip(r)
#vr=np.flip(vr)
plt.plot(r,vr);
plt.xlabel('Clear % of artery');
plt.ylabel('V(r) cc/min');
plt.title('Volume flow vs % blockage');
plt.savefig('flowvblock.png')
#plt.gca().invert_xaxis();
```

```python
mu=4.375359e-8
L=20
V=100
a=.3
```

```python
P = lambda r: 8*mu*L*V*((1-r/100)**(1/2)*a)**-4/np.pi
```

```python
r=np.linspace(1,100,100)
Pr=P(r)
r=r

plt.plot(r,Pr);
plt.xlabel('% blockage of artery');
plt.ylabel('$\Delta P$ mm Hg');
plt.title('Change in Pressure vs % blockage for constant flow');
plt.savefig('pressureblock.png')
#plt.gca().invert_xaxis();
```
